# Kompletny Przewodnik po Pythonie: Co i Jak?

---

## 1. Podstawy (Dla początkujących)

### Witaj, świecie! (`print`)
* **Co robi:** Wyświetla tekst lub wynik działania w konsoli. To Twój sposób komunikacji z programem.

### Zmienne i Typy
* **Zmienne:** Pudełka na dane. 
* **Typy (`int`, `str`, `float`):** Określają, co jest w pudełku (liczba, tekst, ułamek). Python sam rozpoznaje typ.

### Listy
* **Co robią:** Przechowują wiele elementów w jednej zmiennej (np. listę zakupów). Możesz dodawać, usuwać i zmieniać elementy.

### Podstawowe Operatory
* **Co robią:** Wykonują matmę (`+`, `-`, `*`, `/`) oraz operacje logiczne. `%` daje resztę z dzielenia, a `**` potęguje.

### Formatowanie Stringów (f-strings)
* **Co robi:** Pozwala "wstrzykiwać" zmienne prosto do tekstu, np. `f"Masz {punkty} pkt"`.

### Podstawowe Operacje na Stringach
* **Co robią:** Pozwalają zmieniać tekst: powiększać litery (`upper()`), liczyć znaki (`len()`) czy wycinać kawałki tekstu.

### Warunki (`if`, `else`)
* **Co robią:** Pozwalają programowi podejmować decyzje. "Jeśli (if) pada, weź parasol, w przeciwnym razie (else) idź bez".

### Pętle (`for`, `while`)
* **Co robią:** Powtarzają kod wiele razy. `for` idzie po liście, `while` działa, dopóki jakiś warunek jest prawdziwy.

## 1. Zmienne i Podstawowe Typy (Pudełka na dane)
| Element | Co robi? | Przykład |
| :--- | :--- | :--- |
| `x = 5` | **int**: Przechowuje liczby całkowite. | `wiek = 15` |

## 2. Podstawowe Operatory (Matma i Logika)
| Operator | Co robi? | Przykład |
| :--- | :--- | :--- |
| `+`, `-`, `*`, `/` | Wykonuje dodawanie, odejmowanie, mnożenie i dzielenie. | `10 / 2` -> `5.0` |
| `//` | **Dzielenie całkowite**: Odrzuca resztę (wynik to liczba bez przecinka). | `7 // 2` -> `3` |
| `%` | **Modulo**: Podaje tylko resztę z dzielenia. | `7 % 2` -> `1` |
| `**` | **Potęgowanie**: Podnosi liczbę do potęgi. | `2 ** 3` -> `8` |
| `==`, `!=` | Sprawdza, czy wartości są równe lub różne. | `5 == 5` -> `True` |

# Biblioteki do Danych: NumPy i Pandas

---

## 1. NumPy (Numerical Python)
Biblioteka do szybkich obliczeń na dużych zbiorach liczb. Wszystko kręci się wokół obiektu **`array`** (tablica).

| Funkcja / Atrybut | Co robi? | Przykład |
| :--- | :--- | :--- |
| `import numpy as np` | Standardowy sposób importowania biblioteki. | `import numpy as np` |
| `np.array([1, 2, 3])` | Tworzy tablicę NumPy (szybsza niż zwykła lista). | `arr = np.array([1, 2])` |
| `np.zeros(5)` | Tworzy tablicę wypełnioną zerami. | `np.zeros(5)` -> `[0,0,0,0,0]` |
| `np.ones(5)` | Tworzy tablicę wypełnioną jedynkami. | `np.ones(3)` -> `[1,1,1]` |
| `np.arange(0, 10, 2)`| Tworzy ciąg liczb (od, do, krok). | `[0, 2, 4, 6, 8]` |
| `arr.reshape(2, 2)` | Zmienia kształt tablicy (np. z listy na tabelę). | `arr.reshape(2, 5)` |
| `np.mean(arr)` | Oblicza **średnią** z liczb w tablicy. | `np.mean([1, 2, 3])` -> `2.0` |
| `np.median(arr)` | Wyznacza **medianę** (środkową wartość). | `np.median([1, 5, 10])` -> `5.0` |
| `np.dot(a, b)` | Wykonuje mnożenie macierzy. | `np.dot(m1, m2)` |



---

## 2. Pandas (Panel Data)
Biblioteka do analizy danych. Główny obiekt to **`DataFrame`** – czyli tabela z wierszami i kolumnami.

| Funkcja / Metoda | Co robi? | Przykład |
| :--- | :--- | :--- |
| `import pandas as pd` | Standardowy sposób importowania. | `import pandas as pd` |
| `pd.read_csv('f.csv')`| **Wczytuje plik CSV** i zamienia go w tabelę. | `df = pd.read_csv('dane.csv')` |
| `df.head(n)` | Pokazuje `n` pierwszych wierszy tabeli (podgląd). | `df.head(5)` |
| `df.info()` | Wyświetla informacje o kolumnach i brakujących danych. | `df.info()` |
| `df.describe()` | Generuje **statystyki** (średnia, min, max) dla tabeli. | `df.describe()` |
| `df['Kolumna']` | Wybiera jedną konkretną kolumnę z tabeli. | `df['Cena']` |
| `df[df['A'] > 5]` | **Filtruje** wiersze (pokaż tylko gdzie A > 5). | `df[df['wiek'] >= 18]` |
| `df.groupby('X')` | Grupuje dane (np. oblicz średnią sprzedaż na miasto). | `df.groupby('Miasto').sum()` |
| `df.fillna(0)` | Zastępuje brakujące dane (NaN) wybraną wartością. | `df.fillna(0)` |
| `df.to_excel('f.xlsx')`| Zapisuje Twoją tabelę do pliku Excel. | `df.to_excel('wynik.xlsx')` |



---

## 3. Dlaczego ich używamy? (Krótko)

* **NumPy:** Bo jest **ekstremalnie szybki**. Operacje na milionach liczb trwają ułamki sekund, bo są wykonywane bezpośrednio w pamięci procesora (w języku C).
* **Pandas:** Bo ułatwia życie. Zamiast pisać pętle `for`, żeby policzyć coś w tabeli, piszesz jedną komendę (np. `df.mean()`).

---

## 4. Przykład połączenia (Workflow)
W typowym programie robisz to tak:
1. **Pandas:** Wczytujesz dane z pliku.
2. **Pandas:** Czyścisz dane (usuwasz puste wiersze).
3. **NumPy:** Wykonujesz ciężkie obliczenia matematyczne na kolumnach.
4. **Pandas:** Zapisujesz wynik do nowego pliku.
