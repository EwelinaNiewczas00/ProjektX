BRO CODE TUTORIAL - list comprehensions 

```python
doubles =[]
for x in range (1,11):
    doubles.append(x*2)
print(doubles)
```

```python
doubles =[x*2 for x in range(1, 11)]
print(doubles)
```

```python
triples = [y*3 for y in range(1, 11)]
print(triples)
```

```python
owoce =["apple", "orange", "banana"]
uppercase = [owoc.upper() for owoc in owoce]
print(uppercase)
```

```python
uppercase = [owoc.upper() for owoc in ["apple", "orange", "banana"]]
print(uppercase)
```

```python
owoce =["apple", "orange", "banana"]
lista = [owoc[0] for owoc in owoce]
print(lista)
```

```python
number = [1, -2, 4, -6, 9, -6]
pozytywne = [num for num in number if num >=0]
negatywne = [num for num in number if num <0]
inne = [num for num in number if num %2 ==0]

print(pozytywne)
print(negatywne)
print(inne)
```

MAP 

```python
prices = ["$12.50", "$9.50", "$10.50"]
print(list(map(lambda p : float(p.replace("$","")), prices)))

"""Tworze formułe dla 1 wartosci, aby zrobic lambdke dla kazdej zmapować i wyprostować w list
p = "$12.50"
print(float(p.replace("$","")))"""
      
```

```python
liczby = [1, 2, 3, 4, 5]
# map(funkcja, kolekcja)
kwadraty = list(map(lambda x: x**2, liczby))
# Wynik: [1, 4, 9, 16, 25]
print(kwadraty)
```

```python
imiona = ["Ania", "BOB", "Celina"]
male_imiona = list(map(lambda s: s.lower(), imiona))
# Wynik: ['ania', 'bob', 'celina']
print(male_imiona)
```

```python
kody = [101, 102, 103]
nazwy = list(map(lambda x: f"ID_{x}", kody))
# Wynik: ['ID_101', 'ID_102', 'ID_103']
print(nazwy)
```

FILTER

```python
# Lista liczb
numbers = [1, 5, 8, 10, 15, 20, 3]

# Przefiltrowanie: zostawiamy tylko liczby większe od 10
filtered_numbers = list(filter(lambda x: x > 10, numbers))

print(filtered_numbers)  # Wynik: [15, 20]
```

```python
studenci = [["Maria", 90],["Kumar", 85],["Max",60]]

print(list(filter(lambda row: row[1] >70, studenci)))


```

```python
teksty = ["python", "", "lambda", " ", "filter", ""]
niepuste = list(filter(lambda s: s and s.strip(), teksty))
# Wynik: ['python', 'lambda', ' ']
print(niepuste)
```

```python
produkty = [
    {"nazwa": "mleko", "cena": 5},
    {"nazwa": "chleb", "cena": 3},
    {"nazwa": "jajka", "cena": 10}
]
tanie = list(filter(lambda p: p["cena"] < 7, produkty))
# Wynik: [{'nazwa': 'mleko', 'cena': 5}, {'nazwa': 'chleb', 'cena': 3}]
print(tanie)
```

SORT 


```python
values = [(1, "b", "hello"), (2, "a", "wordl"), (2, "c", "ok")]
sorted_value = sorted(values, key=lambda x: x[1])
print(list(sorted_value))
```

```python
ludzie = [{'imie': 'Ania', 'wiek': 25}, {'imie': 'Bartek', 'wiek': 20}]
# Sortowanie według wieku
posortowani = sorted(ludzie, key=lambda x: x['wiek'])
# Wynik: [{'imie': 'Bartek', 'wiek': 20}, {'imie': 'Ania', 'wiek': 25}]
print(posortowani)
```

```python
dane = [('jabłko', 5), ('banan', 2), ('cytryna', 8)]
# Sortowanie według liczby (drugi element, indeks 1)
posortowane = sorted(dane, key=lambda x: x[1])
# Wynik: [('banan', 2), ('jabłko', 5), ('cytryna', 8)]

print(posortowane)
```

```python
slowa = ['dom', 'lokomotywa', 'kod']
# Sortowanie według długości słowa
posortowane = sorted(slowa, key=lambda x: len(x))
# Wynik: ['kod', 'dom', 'lokomotywa']
print(posortowane)
```

```python
# Sortuj po wieku (rosnąco), a potem po imieniu (alfabetycznie)
posortowani = sorted(ludzie, key=lambda x: (x['wiek'], x['imie']))
print(posortowani)
```

REDUCE 

```python
from functools import reduce
nums = [1, 2, 3, 4]
# Sumowanie: ((1+2)+3)+4
total = reduce(lambda x, y: x + y, nums) # Wynik: 10
print(total)
```

```python
from functools import reduce
data = [{'a': 1}, {'b': 2}, {'c': 3}]
# Scalanie słowników
combined = reduce(lambda x, y: {**x, **y}, data) # Wynik: {'a': 1, 'b': 2, 'c': 3}
print(combined)
```

```python
from functools import reduce

nums =[3,2,5,6,3,8,6,9]
sum = reduce(lambda a,b: a+b, nums)
print(sum)
```
