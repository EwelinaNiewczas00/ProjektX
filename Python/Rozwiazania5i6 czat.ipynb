```python
# Zadanie 9 – Filtrowanie outlierów
# Masz listę pomiarów. Outlier to wartość poza [20, 30]. Stwórz listę bez nich (list comprehension).
pomiary = [23, 24, 22, 150, 23, 25, -10, 24, 23]
clean = [x for x in pomiary if 20 <= x <= 30]
print(f"Oczyszczone: {clean}")

# Zadanie 10 – Agregacja danych po kategoriach
# Oblicz łączną kwotę dla każdej kategorii z listy transakcji. Zapisz w słowniku.
trans = [('food', 50), ('transport', 20), ('food', 30), ('entert.', 40), ('transport', 15), ('food', 25)]
sums = {}
for cat, val in trans:
    sums[cat] = sums.get(cat, 0) + val
print(f"Agregacja: {sums}")

# Zadanie 11 – Normalizacja tekstów
# Znormalizuj listę kategorii (strip + lower) i znajdź unikalne (set).
cats = [" Electronics ", "ELECTRONICS", "electronics", " Books", "books ", "BOOKS"]
unique = set(c.strip().lower() for c in cats)
print(f"Unikalne: {unique}")

# Zadanie 12 – Tworzenie SQL query
# Zbuduj query SELECT id, name... używając join().
tabela = "customers"
cols = ['id', 'name', 'email', 'city']
query = f"SELECT {', '.join(cols)} FROM {tabela} WHERE city = 'New York'"
print(query)
```

```python
# Zadanie 13 – Analiza Iris dataset
# Dla gatunku setosa oblicz śr. dł. płatka, max szer. i ratio dł/szer. Wyświetl tabelę.
iris_setosa = [{'l': 1.4, 'w': 0.2}, {'l': 1.5, 'w': 0.3}, {'l': 1.3, 'w': 0.2}]
avg_l = sum(d['l'] for d in iris_setosa) / len(iris_setosa)
max_w = max(d['w'] for d in iris_setosa)
print(f"{'Gatunek':<10} | {'Avg L':<6} | {'Max W':<6} | {'Ratio':<5}")
print(f"{'setosa':<10} | {avg_l:<6.2f} | {max_w:<6.2f} | {avg_l/max_w:<5.2f}")

# Zadanie 14 – Ranking modeli
# Oblicz efficiency score = accuracy / log10(czas). Posortuj i wyświetl top 3.
import math
models = [('LR', 0.78, 1.2), ('RF', 0.92, 45.6), ('SVM', 0.85, 12.3), ('XGB', 0.95, 67.8)]
ranked = sorted([(m[0], m[1]/math.log10(m[2])) for m in models], key=lambda x: x[1], reverse=True)
print(f"Ranking: {ranked[:3]}")

# Zadanie 15 – Parsowanie logów
# Wyciągnij i zlicz poziomy logów (INFO, WARNING, ERROR). Oblicz %.
logs = ["... INFO ...", "... WARNING ...", "... ERROR ...", "... INFO ..."]
lvls = [l.split(" ")[1] for l in logs]
for v in set(lvls):
    print(f"{v}: {lvls.count(v)} ({lvls.count(v)/len(lvls):.0%})")

# Zadanie 16 – Confusion Matrix z list
# Oblicz TP, TN, FP, FN oraz accuracy i precision z pętli.
y_t, y_p = [1, 0, 1, 1, 0], [1, 0, 1, 0, 0]
tp = sum(1 for t, p in zip(y_t, y_p) if t == 1 and p == 1)
fp = sum(1 for t, p in zip(y_t, y_p) if t == 0 and p == 1)
# ... pozostałe analogicznie
print(f"TP: {tp}, FP: {fp}, Precision: {tp/(tp+fp) if (tp+fp)>0 else 0}")

# Zadanie 17 – Czyszczenie danych CSV
# Usuń białe znaki, zamień puste na "Unknown", stwórz listę słowników.
raw_csv = ["1,Alice,28, NY ", "2, Bob ,, Boston"]
data = []
for line in raw_csv:
    vals = [v.strip() if v.strip() else "Unknown" for v in line.split(",")]
    data.append({"id": vals[0], "name": vals[1], "city": vals[3]})
print(data)

# Zadanie 18 – Agregacja z grupowaniem
# Oblicz sumę, średnią i count dla kategorii w sprzedaży.
sales = [('Laptop', 'Elec', 1200), ('Phone', 'Elec', 800), ('Book', 'Media', 15)]
agg = {}
for prod, cat, price in sales:
    agg.setdefault(cat, []).append(price)
for c, p in agg.items():
    print(f"{c}: Sum={sum(p)}, Avg={sum(p)/len(p)}, Count={len(p)}")

# Zadanie 19 – Generowanie raportu markdown
# Stwórz tabelę MD z danymi eksperymentów używając join.
exps = [{'n': 'E1', 'a': 0.85, 'p': 0.83}]
head = "| Exp | Acc | Prec |\n|---|---|---|"
rows = [f"| {e['n']} | {e['a']:.1%} | {e['p']:.1%} |" for e in exps]
print(head + "\n" + "\n".join(rows))

# Zadanie 20 – Walidacja i transformacja pipeline
# Funkcja clean_and_validate: strip, try/except float, zakres [0,1].
def validate(raw):
    valid, errors = [], []
    for i, v in enumerate(raw):
        try:
            val = float(v.strip())
            if 0 <= val <= 1: valid.append(val)
            else: errors.append((i, v, "Range"))
        except: errors.append((i, v, "Type"))
    return valid, errors
print(validate([" 0.8 ", "bad", "1.5"]))
```

```python
# Zadanie 9 – Inwersja słownika
# Stwórz odwrotny słownik: kategorie jako klucze, listy produktów jako wartości.
produkty_kat = {'laptop': 'elektronika', 'klawiatura': 'elektronika', 'krzesło': 'meble', 'biurko': 'meble'}
kat_produkty = {}
for prod, kat in produkty_kat.items():
    kat_produkty.setdefault(kat, []).append(prod)
print(f"Inwersja: {kat_produkty}")

# Zadanie 10 – Walidacja danych ze zbiorem
# Znajdź nieprawidłowe kategorie używając operacji zbiorowych.
user_cats = ['sport', 'muzyka', 'SPORT', 'film', 'xyz', 'nauka', 'gaming']
allowed = {'sport', 'muzyka', 'film', 'nauka', 'gry'}
cleaned_user_cats = {c.lower() for c in user_cats}

nieprawidlowe = cleaned_user_cats - allowed
prawidlowe = cleaned_user_cats & allowed
print(f"Prawidłowe: {prawidlowe}\nNieprawidłowe: {nieprawidlowe}")

# Zadanie 11 – Dictionary comprehension
# Słownik kwadratów 1-20, a potem filtrowanie parzystych.
kwadraty = {x: x**2 for x in range(1, 21)}
parzyste_kwadraty = {k: v for k, v in kwadraty.items() if k % 2 == 0}
print(f"Wszystkie: {kwadraty}\nParzyste: {parzyste_kwadraty}")

# Zadanie 12 – Analiza transakcji - słowniki zagnieżdżone
# Stwórz strukturę {data: {kategoria: suma}}.
transakcje = [('2024-01-15', 'Elektronika', 1200), ('2024-01-15', 'Meble', 450), ('2024-01-16', 'Elektronika', 800), ('2024-01-16', 'Elektronika', 350), ('2024-01-15', 'Odzież', 200), ('2024-01-16', 'Meble', 600)]
analiza = {}
for data, kat, kwota in transakcje:
    analiza.setdefault(data, {})
    analiza[data][kat] = analiza[data].get(kat, 0) + kwota
print(f"Analiza: {analiza}")
```

```python
# Zadanie 13 – Analiza sentymentu
komentarze = ["Ten produkt jest świetny i fantastyczny", "Okropny towar, kiepska jakość", "Całkiem dobry, ale mogłoby być lepiej", "Rewelacja, super jakość, polecam", "Słaby produkt, nie polecam"]
pos, neg = {'świetny', 'fantastyczny', 'dobry', 'super', 'rewelacja', 'polecam'}, {'okropny', 'kiepski', 'słaby', 'zły'}

analiza = []
for k in komentarze:
    slowa = k.lower().replace(",", "").split()
    p_count = len([s for s in slowa if s in pos])
    n_count = len([s for s in slowa if s in neg])
    # Obsługa "nie polecam" jako frazy (uproszczona)
    if "nie polecam" in k.lower(): n_count += 1
    sent = "pozytywny" if p_count > n_count else "negatywny" if n_count > p_count else "neutralny"
    analiza.append((k, sent, p_count, n_count))

for res in analiza: print(res)

# Zadanie 14 – Grupowanie studentów według średniej
studenci = [('Anna', 'Kowalska', 5.0, 4.5, 5.0), ('Jan', 'Nowak', 3.0, 3.5, 3.0), ('Maria', 'Zielińska', 4.0, 4.5, 5.0), ('Piotr', 'Wiśniewski', 2.0, 3.0, 2.5), ('Ewa', 'Wójcik', 5.0, 5.0, 4.5), ('Tomasz', 'Kamiński', 4.0, 3.5, 4.0)]
grupy = {'wyróżniający': [], 'dobry': [], 'średni': []}
for s in studenci:
    sr = sum(s[2:]) / 3
    kat = 'wyróżniający' if sr >= 4.5 else 'dobry' if sr >= 3.5 else 'średni'
    grupy[kat].append((s[0], s[1], round(sr, 2)))
print(grupy)

# Zadanie 15 – Przecięcia w dniach aktywności
d1 = {'user001', 'user002', 'user003', 'user004', 'user005', 'user010'}
d2 = {'user002', 'user003', 'user006', 'user007', 'user008', 'user010'}
d3 = {'user003', 'user004', 'user009', 'user010', 'user011'}

wszystkie = d1 & d2 & d3
tylko_jeden = (d1 - d2 - d3) | (d2 - d1 - d3) | (d3 - d1 - d2)
dokladnie_dwa = ((d1 & d2) | (d2 & d3) | (d1 & d3)) - wszystkie
print(f"3 dni: {wszystkie}\nTylko 1 dzień: {tylko_jeden}\nDokładnie 2 dni: {dokladnie_dwa}")

# Zadanie 16 – Fibonacci z Cache (Memoization)
cache = {}
def fibonacci(n):
    if n in cache: return cache[n]
    if n <= 1: return n
    cache[n] = fibonacci(n-1) + fibonacci(n-2)
    return cache[n]

import time
s = time.time()
print(f"Fib(35): {fibonacci(35)}, Czas: {time.time()-s:.5f}s")

# Zadanie 17 – Inwersja many-to-many (filmy i gatunki)
filmy = {'Matrix': ['Akcja', 'Sci-Fi'], 'Shrek': ['Komedia', 'Familijny'], 'Inception': ['Akcja', 'Sci-Fi', 'Thriller']}
gatunki = {}
for f, g_list in filmy.items():
    for g in g_list:
        gatunki.setdefault(g, set()).add(f)
print(gatunki)

# Zadanie 18 – Podobieństwo Jaccarda
def jaccard(s1, s2):
    return len(s1 & s2) / len(s1 | s2)

k1 = {'mleko', 'chleb', 'masło', 'jajka', 'ser'}
k2 = {'mleko', 'chleb', 'dżem', 'jogurt', 'ser'}
print(f"Jaccard (1,2): {jaccard(k1, k2):.2f}")

# Zadanie 19 – Top N słów (Word frequency)
tekst = "Python jest językiem programowania który jest bardzo popularny...".lower()
stop = {'jest', 'do', 'i', 'w', 'można', 'go'}
words = [w for w in tekst.replace(".", "").split() if w not in stop]
freq = {w: words.count(w) for w in set(words)}
top10 = sorted(freq.items(), key=lambda x: x[1], reverse=True)[:10]
print(f"Top 10: {top10}")

# Zadanie 20 – Deduplikacja z zachowaniem kolejności
uzytkownicy = [('user001', 'A', 'K'), ('user002', 'J', 'N'), ('user001', 'A', 'K'), ('user003', 'M', 'Z')]
seen = set()
clean = []
for u in uzytkownicy:
    if u[0] not in seen:
        clean.append(u)
        seen.add(u[0])
print(f"Oryginalna kolejność bez duplikatów: {clean}")
```
