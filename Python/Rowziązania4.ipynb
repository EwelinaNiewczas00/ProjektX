1. ✏ Zadanie 1 – Kalkulator BMI
Napisz program, który obliczy wskaźnik BMI (Body Mass Index).
Dane:
Waga: 70 kg
Wzrost: 1.75 m
Wymagania:
Stwórz zmienne waga i wzrost
Oblicz BMI według wzoru: BMI = waga / (wzrost²)
Wyświetl wynik z 2 miejscami po przecinku
Dodaj interpretację: BMI < 18.5 (niedowaga), 18.5-24.9 (norma), ≥25 (nadwaga)
Oczekiwany rezultat:
BMI: 22.86
Interpretacja: Norma

```python
waga = 70
wzrost = 1.75

BMI = waga / (wzrost ** 2)

if BMI < 18.5:
    x = "niedowaga"
elif 18.5 <= BMI <= 24.9:
    x = "norma"
elif BMI >= 25:
    x = "nadwaga"

print(f"BMI: {BMI:.2f} \nInterpretacja: {x.capitalize()}")
```

9. ✏ Zadanie 9 – Normalizacja Min-Max
Znormalizuj wartość do zakresu [0, 1].
Dane:
Wartość: 75
Minimum w kolumnie: 50
Maximum w kolumnie: 100
Wymagania:
Wzór: normalized = (value - min) / (max - min)
Oblicz dla wartości 75, 50, 100, 62.5
Sprawdź, czy wynik mieści się w [0, 1]
Wyświetl tabelkę: wartość oryginalna | wartość znormalizowana

```python
def normalize(val, v_min, v_max):
    return (val - v_min) / (v_max - v_min)

v_min, v_max = 50, 100
test_values = [75, 50, 100, 62.5]

print(f"{'Oryginał':<10} | {'Znormalizowana':<15}")
print("-" * 30)
for v in test_values:
    norm = normalize(v, v_min, v_max)
    print(f"{v:<10.1f} | {norm:<15.2f}")
```

12. ✏ Zadanie 12 – Estymacja czasu treningu
Oblicz szacowany czas treningu modelu.
Dane:
Liczba próbek: 50000
Batch size: 128
Liczba epok: 50
Czas na jeden batch: 0.15 sekundy
Wymagania:
Oblicz liczbę batchy na epokę (użyj dzielenia całkowitego i modulo)
Oblicz całkowity czas w sekundach
Skonwertuj na minuty i godziny
Wyświetl: "Szacowany czas: X h Y min Z s"

```python
samples = 50000
batch_size = 128
epochs = 50
time_per_batch = 0.15

batches_per_epoch = (samples // batch_size) + (1 if samples % batch_size != 0 else 0)
total_seconds = int(batches_per_epoch * epochs * time_per_batch)

h = total_seconds // 3600
m = (total_seconds % 3600) // 60
s = total_seconds % 60

print(f"Szacowany czas: {h} h {m} min {s} s")
```

13. 🧠 Zadanie 13 – Detekcja anomalii prostą metodą
Wykryj wartości odstające używając reguły 3-sigma.
Dane (5 pomiarów ciśnienia krwi):
120, 125, 118, 190, 122 mmHg
Wymagania:
Oblicz średnią i odchylenie standardowe (ręcznie, bez bibliotek)
Zidentyfikuj wartości, które odchylają się o więcej niż 3×std od średniej
Oznacz je jako anomalie
Wyświetl raport z wszystkimi wartościami i oznaczeniem anomalii
Oczekiwany rezultat:
Wartość: 120 - OK
Wartość: 125 - OK
Wartość: 118 - OK
Wartość: 190 - ANOMALIA!
Wartość: 122 - OK

```python
data = [120, 125, 118, 190, 122]
mean = sum(data) / len(data)
variance = sum((x - mean)**2 for x in data) / len(data)
std = variance ** 0.5

limit_up = mean + 3 * std
limit_down = mean - 3 * std

for x in data:
    status = "ANOMALIA!" if x > limit_up or x < limit_down else "OK"
    print(f"Wartość: {x} - {status}")
```

15. 🧠 Zadanie 15 – Standaryzacja Z-score
Standaryzuj wartości używając wzoru z-score.
Dane (wagi pacjentów w kg):
65.5, 78.2, 54.8, 92.1, 70.0
Wymagania:
Oblicz średnią i odchylenie standardowe
Dla każdej wartości oblicz z-score: z = (x - mean) / std
Sprawdź, że średnia z-scores ≈ 0 i odchylenie ≈ 1
Wyświetl oryginalną wartość i jej z-score
Zinterpretuj: z > 2 (wartość bardzo wysoka), |z| < 1 (typowa), z < -2 (bardzo niska)

```python
weights = [65.5, 78.2, 54.8, 92.1, 70.0]
mean = sum(weights) / len(weights)
std = (sum((x - mean)**2 for x in weights) / len(weights)) ** 0.5

for x in weights:
    z = (x - mean) / std
    if z > 2: interp = "bardzo wysoka"
    elif abs(z) < 1: interp = "typowa"
    elif z < -2: interp = "bardzo niska"
    else: interp = "średnia"
    print(f"Waga: {x:<5} | Z-score: {z:>6.2f} | Interpretacja: {interp}")
```

19. 🧠 Zadanie 19 – Kalkulator accuracy z threshold-em
Oblicz accuracy dla różnych progów klasyfikacji.
Dane (prawdopodobieństwa i prawdziwe etykiety):
Prawdopodobieństwa: 0.23, 0.67, 0.89, 0.34, 0.91, 0.12, 0.78, 0.45
Prawdziwe klasy: 0, 1, 1, 0, 1, 0, 1, 0
Wymagania:
Dla threshold = 0.5: jeśli prob >= 0.5 → klasa 1, inaczej klasa 0
Porównaj predykcje z prawdziwymi klasami
Oblicz accuracy = (liczba poprawnych) / (liczba wszystkich)
Powtórz dla threshold = 0.3, 0.5, 0.7
Wyświetl accuracy dla każdego threshold i wskaż optymalny

```python
probs = [0.23, 0.67, 0.89, 0.34, 0.91, 0.12, 0.78, 0.45]
labels = [0, 1, 1, 0, 1, 0, 1, 0]

for th in [0.3, 0.5, 0.7]:
    preds = [1 if p >= th else 0 for p in probs]
    correct = sum(1 for p, l in zip(preds, labels) if p == l)
    acc = correct / len(labels)
    print(f"Threshold: {th} -> Accuracy: {acc:.2%}")
```
